```python
#importing dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from sklearn.metrics import accuracy_score,classification_report
```

```python
#importing the iris dataset
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
iris_dataset
```

```python
dataset = pd.DataFrame(iris_dataset.data,columns=iris_dataset.feature_names)
```

```python
dataset
```

```python
dataset['label'] = iris_dataset.target
```

```python
dataset
```

```python
#getting some info about the dataset
dataset.info()
```

```python
dataset['label'].value_counts()
```

```python
#checking the null values
dataset.isnull().sum()
```

```python
#getting some statistical measures of the dataset
dataset.describe()
```

```python
X = dataset.drop(columns='label',axis=1)
Y = dataset['label']
```

```python
#splitting the dataset into train-test-split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
#Preprocessing the dataset using the standard scaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

```python
X.shape,X_train_std.shape,X_test_std.shape
```

```python
#making a ann model
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(4,)))
model.add(keras.layers.Dense(10,activation='relu'))
model.add(keras.layers.Dense(15,activation='relu'))
model.add(keras.layers.Dense(3,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss = 'sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=30)
```

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)
print("Accuracy for testing data is ",round(accuracy*100,2),"%")
```
