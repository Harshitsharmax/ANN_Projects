**Breast Cancer Classification using NN**

I have developed a robust breast cancer classification model using deep neural networks, focusing on numerical data such as tumor features like mean radius, texture, and other clinical parameters. Utilizing deep learning techniques, the model effectively processes and interprets these numerical features to accurately classify tumors as malignant or benign. Trained on a comprehensive dataset of patient records and tumor characteristics, the neural network achieves high performance in predicting breast cancer outcomes, providing valuable support for clinical decision-making and enhancing diagnostic accuracy.



```python
# Importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Loading the data from Sklearn
dataset = sklearn.datasets.load_breast_cancer()
print(dataset)
```

```python
# Loading the dataset to pandas dataframe
data_frame = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
# Checking the first five rows of the data frame
data_frame.head()
```

```python
# Adding the target column to the data frame
data_frame['label'] = dataset.target
```

```python
# Print the last 5 rows of the data frame
data_frame.tail()
```

```python
# Checking the number of rows and columns of the data frame
data_frame.shape
```

```python
# Getting some information about the data
data_frame.info()
```

```python
# Checking for any missing values
data_frame.isnull().sum()
```

```python
# Getting the statistical measures of the data
data_frame.describe()
```

```python
# Checking the distribution of the target variable
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
# Checking the mean vakue for each of the label
data_frame.groupby('label').mean
```

```python
# Separating the features and the target
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
# Splitting the data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)
```

```python
# Standardizing the data
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

**Building the Neural Network**

```python
# Importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
# Compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the Neural Network
hs = model.fit(x_train_std,y_train,validation_split = 0.1,epochs = 20)
```

```python
# Visualizing accuracy
plt.plot(hs.history['accuracy'])
plt.plot(hs.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc = 'lower right')
```

```python
# Visualizing loss
plt.plot(hs.history['loss'])
plt.plot(hs.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc = 'upper right')
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```

**Building the Predictive system**

```python
# Taking input from the user
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# Change the input to a numpy array
input_array = np.asarray(input_data)

# Reshaping the array as we are predicting for only one data point
array_reshaped = input_array.reshape(1,-1)

# Standardizing the input data
input_std = scaler.transform(array_reshaped)

# Making Prediction
prediction = model.predict(input_std)
prediction_label = [np.argmax(prediction)]

if (prediction_label[0] == 0):
  print("The tumor is Malignant")
else:
  print("The tumor is Benign")
```
